```
list_firstNames=[]
list_lastNames=[]
list_fullNames=[]
filepath = 'test-data-10-exp-5.list'
with open(filepath) as fp:
   line = fp.readline()
   cnt = 1
   while line:
       #print("Line {}: {}".format(cnt, line.strip()))
       line = fp.readline()
       if '--' in line:
          a=line.find(',')
          b=line.find('-')
          last_names = line[0:a]
          first_names = line[a+2:b]
          full_names = line[0:b]
          list_lastNames.append(last_names)
          list_firstNames.append(first_names)
          list_fullNames.append(full_names)
       cnt += 1
   print(cnt)
```
